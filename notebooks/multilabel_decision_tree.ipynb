```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.datasets import make_multilabel_classification
import graphviz
# from dtreeviz.trees import dtreeviz
```

```python
# サンプルデータの生成
X, y = make_multilabel_classification(
    n_samples=100,
    n_features=10,
    n_classes=5,
    n_labels=5,
    random_state=0
)
```

```python
# データをPandas Dataframeに変換し、特徴量名とクラス名を生成
import string
alphabets = list(string.ascii_lowercase)

cols = [f"col_{alphabets[i]}" for i in range(X.shape[1])]
X = pd.DataFrame(X, columns=cols)

labels = [f"label_{alphabets[i].upper()}" for i in range(y.shape[1])]
```

```python
# 決定木で学習
model = DecisionTreeClassifier(random_state=0, max_depth=5)
model.fit(X, y)
```

## GraphViz

```python
dot = export_graphviz(
    model,
    filled=True,
    rounded=True,
    class_names=labels,
    feature_names=cols,
    out_file=None
)
graph = graphviz.Source(dot)
```
